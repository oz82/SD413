```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[6000], [8200], [9000], [14200], [16200]]).reshape(-1, 1)
y = [86000, 82000, 78000, 75000, 70000]

# grafiğin çizilmesi
plt.figure()
plt.title('Otomobil Fiyat-KM Dağılım Grafiği')
plt.xlabel('KM')
plt.ylabel('Fiyat')
plt.plot(x, y, 'k.')
plt.axis([3000, 20000, 60000, 95000])
plt.grid(True)
plt.show()

```

```python
# modelin oluşturulması
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
print(model.intercept_)
print(model.coef_)
```

```python
plt.figure()
plt.title('Otomobil Fiyat-KM Dağılım Grafiği')
plt.xlabel('KM')
plt.ylabel('Fiyat')
plt.plot(x, model.predict(x), color = 'red')
plt.plot(x, y, 'k.')
plt.grid(True)
plt.show()
```

```python
test = np.array([[9000]])
price = model.predict(test)[0]
print(price)
```

```python
y_predictions = model.predict(x)
for i, prediction in enumerate(y_predictions):
    print('tahmin: %.2f, gerçek: %s' %(prediction, y[i]))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print(mean_absolute_error(y, y_predictions))
print(mean_squared_error(y, y_predictions))
print(r2_score(y, y_predictions))
```
