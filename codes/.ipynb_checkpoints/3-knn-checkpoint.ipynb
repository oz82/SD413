```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([
    [160,65],[170,85],[185,85],[188,82],[155,50],
    [161,58],[180,68],[157,52],[170,66]])
y_train = ['K', 'E', 'E', 'E', 'K', 'K', 'E', 'K', 'K']

plt.figure(figsize=(12,7))
plt.title('Cinsiyete göre Boy ve Ağırlık')
plt.xlabel('Boy (cm)')
plt.ylabel('Ağırlık (kg)')
for i, x in enumerate(x_train):
    plt.scatter(x[0], x[1], c='k', marker='x' if y_train[i]=='E' else 'D')
plt.grid(True)
plt.show()

from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(y_train)

from sklearn import neighbors
n_neighbors = 3
clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
clf.fit(x_train, y_train)
clf.predict(x_train)

test = np.array([
    [200,110],[180,90],[145,45],[195,55],[155,100]])
clf.predict(test)
```
